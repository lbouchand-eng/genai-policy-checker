```python
import os
from getpass import getpass

os.environ["OPENAI_API_KEY"] = os.getenv("OPENAI_API_KEY") \
    or getpass("Enter your OpenAI API key: ")
```

```python
# ========================
# üìò Create Vector Store (Chroma)
# ========================
import json
from tqdm import tqdm
from langchain_openai import OpenAIEmbeddings
from langchain_community.vectorstores import Chroma
from langchain_openai import OpenAIEmbeddings


# ---- CONFIG ----
CHUNKS_FILE = "eu_laws_chunks.jsonl"
CHROMA_DIR = "chroma_eu_laws"
# ----------------

# 1Ô∏è‚É£ Charger les chunks
chunks = []
with open(CHUNKS_FILE, "r", encoding="utf-8") as f:
    for line in f:
        chunks.append(json.loads(line))

print(f"Loaded {len(chunks)} chunks")

# 2Ô∏è‚É£ Construire les documents LangChain
from langchain_core.documents import Document

docs = [
    Document(
        page_content=c["text"],
        metadata={
            "source": c["document_title"],
            "section": c.get("section"),
            "article": c.get("article_number"),
            "recital": c.get("number")
        }
    )
    for c in chunks
]

# 3Ô∏è‚É£ Cr√©er les embeddings
embeddings = OpenAIEmbeddings(model="text-embedding-3-small")
# 4Ô∏è‚É£ Cr√©er / persister la base Chroma
vectorstore = Chroma.from_documents(docs, embeddings, persist_directory=CHROMA_DIR)
vectorstore.persist()

print(f"‚úÖ Vector store created and saved in {CHROMA_DIR}")
```
